```python
import requests
import pandas as pd
import json
import xml.etree.ElementTree as ET 

def fetch(url):
    r = requests.get(url)
    data = r.json()
    return data
  
def save(data, filename):
    with open(filename, 'w') as f:
        json.dump(data, f)

def load(filename):
    with open(filename, 'r') as f:
        data = json.load(f)
    return data 
```

```python
appended_data = []
df = pd.read_csv('./dataset_w_weather&rwi.csv');
display(df)
print(len(df))
df['city'] = ''
for index, row in df.iterrows():
  # print progress bar
  print("{0}/{1}".format(index, len(df)))
  lat = row['latitude']
  lng = row['longitude']
  
  url = "https://nominatim.openstreetmap.org/reverse?lat={0}&lon={1}".format(lat, lng)
  r = requests.get(url)
  root = ET.fromstring(r.content) 
  for child in root:
    if child.tag == "addressparts":
      for subchild in child:
        if subchild.tag == "city":
          df.at[index, 'city'] = subchild.text
          break

df
```
