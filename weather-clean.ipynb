```python
import pandas as pd
import json
```

```python
# load json into dict
with open('./scraper/weather.json') as f:
    data = json.load(f)

data = data
# ori_df = pd.DataFrame(data)
df = pd.DataFrame(columns=['date', 'hour', 'desc'])

# loop data dict
for month in data:
  for hour in month:
    if 'hourly' not in hour:
      break;
    hourly = hour['hourly']
    for everyHour in hourly:
      df = df.append({'date': hour['date'], 'hour': everyHour['time'], 'desc': everyHour['weatherDesc'][0]['value']}, ignore_index=True)
df
```

```python
df.to_csv('weather.csv')
```
